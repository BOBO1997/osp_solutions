```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from pprint import pprint
import pickle
import time
import datetime
%matplotlib inline
```

```python
with open("e2d1_raw.pkl", "rb") as f:
    raw = pickle.load(f)
with open("e2d1_qrem.pkl", "rb") as f:
    qrem = pickle.load(f)
with open("e2d1_zne.pkl", "rb") as f:
    zne = pickle.load(f)
with open("e2d1_zne_pt.pkl", "rb") as f:
    zne_pt = pickle.load(f)
```

```python
num_steps_list = raw["num_steps_list"]
raw_fid_list = raw["fid"]
raw_stddev_list = raw["stddev"]
```

```python
num_steps_list = qrem["num_steps_list"]
qrem_fid_list = qrem["fid"]
qrem_stddev_list = qrem["stddev"]
```

```python
num_steps_list = zne["num_steps_list"]
zne_fid_list = zne["fid"]
zne_stddev_list = zne["stddev"]
```

```python
num_steps_list = zne_pt["num_steps_list"]
zne_pt_fid_list = zne_pt["fid"]
zne_pt_stddev_list = zne_pt["stddev"]
```

```python
plt.clf()
plt.figure(dpi=200)
plt.scatter(num_steps_list, raw_fid_list)
plt.scatter(num_steps_list, qrem_fid_list)
plt.scatter(num_steps_list, zne_fid_list)
plt.scatter(num_steps_list, zne_pt_fid_list)
```

```python
limit = 13
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(num_steps_list[3:limit], raw_fid_list[3:limit], marker="o")
plt.plot(num_steps_list[3:limit], raw_fid_list[3:limit],linewidth=1,linestyle="dotted")
p2 = plt.scatter(num_steps_list[3:limit], qrem_fid_list[3:limit], marker="v")
plt.plot(num_steps_list[3:limit], qrem_fid_list[3:limit],linewidth=1,linestyle="dashed")
p3 = plt.scatter(num_steps_list[3:limit], zne_fid_list[3:limit], marker="x")
plt.plot(num_steps_list[3:limit], zne_fid_list[3:limit],linewidth=1,linestyle="dashdot")
p4 = plt.scatter(num_steps_list[3:limit], zne_pt_fid_list[3:limit], marker="*")
plt.plot(num_steps_list[3:limit], zne_pt_fid_list[3:limit],linewidth=1,linestyle="dotted")
plt.xlabel("number of trotter steps")
plt.ylabel("state fidelity")
plt.legend([p1,p2,p3,p4],
           ["Without QEM", "QREM", "QREM, ZNE", "QREM, ZNE, Twirling"])
plt.ylim(bottom=-0.04,top=1.04)
# plt.title("The effect of error mitigations")
```

```python
limit = 13
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(num_steps_list[3:limit], 1 - np.array(raw_fid_list[3:limit]), marker="o")
plt.plot(num_steps_list[3:limit], 1 - np.array(raw_fid_list[3:limit]),linewidth=1,linestyle="dotted")
p2 = plt.scatter(num_steps_list[3:limit], 1 - np.array(qrem_fid_list[3:limit]), marker="v")
plt.plot(num_steps_list[3:limit], 1 - np.array(qrem_fid_list[3:limit]),linewidth=1,linestyle="dashed")
p3 = plt.scatter(num_steps_list[3:limit], 1 - np.array(zne_fid_list[3:limit]), marker="x")
plt.plot(num_steps_list[3:limit], 1 - np.array(zne_fid_list[3:limit]),linewidth=1,linestyle="dashdot")
p4 = plt.scatter(num_steps_list[3:limit], 1 - np.array(zne_pt_fid_list[3:limit]), marker="*")
plt.plot(num_steps_list[3:limit], 1 - np.array(zne_pt_fid_list[3:limit]),linewidth=1,linestyle="dotted")
plt.xlabel("number of Trotter steps")
plt.ylabel("state infidelity")
plt.ylim(bottom=7.0 * 1e-3,top=1.3 * 1e-0)
plt.legend([p1,p2,p3,p4],
           ["Without QEM", "QREM", "QREM, ZNE", "QREM, ZNE, Twirling"])
# plt.ylim(bottom=-0.04,top=1.04)
plt.yscale("log")
# plt.title("The effect of error mitigations")
```
