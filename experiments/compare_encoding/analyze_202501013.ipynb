```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from pprint import pprint
import pickle
import time
import datetime
%matplotlib inline
```

```python
with open("e0d0_qrem.pkl", "rb") as f:
    e0d0_qrem = pickle.load(f)
with open("e2d1_qrem.pkl", "rb") as f:
    e2d1_qrem = pickle.load(f)
with open("e2d2_qrem.pkl", "rb") as f:
    e2d2_qrem = pickle.load(f)
```

```python
num_steps_list = e0d0_qrem["num_steps_list"]
e0d0_fid_list = e0d0_qrem["fid"]
e0d0_stddev_list = e0d0_qrem["stddev"]
```

```python
num_steps_list = e2d1_qrem["num_steps_list"]
e2d1_fid_list = e2d1_qrem["fid"]
e2d1_stddev_list = e2d1_qrem["stddev"]
```

```python
num_steps_list = e2d2_qrem["num_steps_list"]
e2d2_fid_list = e2d2_qrem["fid"]
e2d2_stddev_list = e2d2_qrem["stddev"]
```

```python
plt.clf()
plt.figure(dpi=200)
plt.scatter(num_steps_list, e0d0_fid_list)
plt.scatter(num_steps_list, e2d1_fid_list)
plt.scatter(num_steps_list, e2d2_fid_list)
```

```python
limit = 13
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(num_steps_list[3:limit], e0d0_fid_list[3:limit], marker="o")
plt.plot(num_steps_list[3:limit], e0d0_fid_list[3:limit],linewidth=1,linestyle="dotted")
p2 = plt.scatter(num_steps_list[3:limit], e2d1_fid_list[3:limit], marker="v")
plt.plot(num_steps_list[3:limit], e2d1_fid_list[3:limit],linewidth=1,linestyle="dashed")
# plt.scatter(num_steps_list[:limit], e2d2_fid_list[:limit], marker="x")
plt.xlabel("number of trotter steps")
plt.ylabel("state fidelity")
plt.ylim(bottom=-0.04,top=1.04)
plt.legend([p1,p2],
           ["general encoding, general decoding", "shallow encoding, specific decoding"])
# plt.title("The effect of encoding and decoding strategy")
```

```python
limit = 13
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(num_steps_list[3:limit], 1-np.array(e0d0_fid_list[3:limit]), marker="o")
plt.plot(num_steps_list[3:limit], 1-np.array(e0d0_fid_list[3:limit]),linewidth=1,linestyle="dotted")
p2 = plt.scatter(num_steps_list[3:limit], 1-np.array(e2d1_fid_list[3:limit]), marker="v")
plt.plot(num_steps_list[3:limit], 1-np.array(e2d1_fid_list[3:limit]),linewidth=1,linestyle="dashed")
# plt.scatter(num_steps_list[:limit], e2d2_fid_list[:limit], marker="x")
plt.xlabel("number of Trotter steps")
plt.ylabel("state infidelity")
# plt.ylim(bottom=-0.04,top=1.04)
plt.yscale("log")
plt.legend([p1,p2],
           ["general encoding, general decoding", "shallow encoding, specific decoding"])
# plt.title("The effect of encoding and decoding strategy")
```
