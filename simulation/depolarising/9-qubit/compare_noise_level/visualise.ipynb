```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from pprint import pprint
import pickle
import time
import datetime
%matplotlib inline
from setting import *
```

```python
with open("run_trotter.pkl", "rb") as f:
    run_trotter = pickle.load(f)
with open("run_proposed.pkl", "rb") as f:
    run_proposed = pickle.load(f)
```

```python
noiselevels_to_fidelities_trotter = run_trotter["noiselevels_to_fidelities"]
noiselevels_to_nums_cnots_trotter = run_trotter["noiselevels_to_nums_cnots"]
noiselevels_to_infidelities_trotter = 1 - np.array(noiselevels_to_fidelities_trotter)
noiselevels_to_infidelities_trotter_min = np.min(noiselevels_to_infidelities_trotter, axis=1)
noiselevels_to_infidelities_trotter_min.shape
```

```python
noiselevels_to_fidelities_proposed = run_proposed["noiselevels_to_fidelities"]
noiselevels_to_nums_cnots_proposed = run_proposed["noiselevels_to_nums_cnots"]
noiselevels_to_infidelities_proposed = 1 - np.array(noiselevels_to_fidelities_proposed)
noiselevels_to_infidelities_proposed_min = np.min(noiselevels_to_infidelities_proposed, axis=1)
len(noiselevels_to_infidelities_proposed_min)
noiselevels_to_infidelities_proposed_min.shape
```

```python
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(ps_dep1[1:], noiselevels_to_infidelities_trotter_min[1:], marker="o")
plt.plot(ps_dep1[1:], noiselevels_to_infidelities_trotter_min[1:],linewidth=1, linestyle='dotted')
p2 = plt.scatter(ps_dep1[1:], noiselevels_to_infidelities_proposed_min[1:], marker="v")
plt.plot(ps_dep1[1:], noiselevels_to_infidelities_proposed_min[1:],linewidth=1, linestyle='dashed')
# plt.scatter(num_steps_list[:limit], e2d2_fidelities[:limit], marker="x")
plt.xlabel("single-qubit depolarising probability")
plt.ylabel("state infidelity")
plt.xscale("log")
plt.yscale("log")
plt.legend([p1, p2], ["Trotter", "proposed"]) # , "shallow encoding and shallow decoding"))
plt.show()
# plt.title("The effect of encoding and decoding strategy")
```

```python
plt.clf()
plt.figure(dpi=200)
p1 = plt.scatter(ps_dep1[3:], noiselevels_to_infidelities_trotter_min[3:], marker="o")
plt.plot(ps_dep1[3:], noiselevels_to_infidelities_trotter_min[3:],linewidth=1, linestyle='dotted')
p2 = plt.scatter(ps_dep1[3:], noiselevels_to_infidelities_proposed_min[3:], marker="v")
plt.plot(ps_dep1[3:], noiselevels_to_infidelities_proposed_min[3:],linewidth=1, linestyle='dashed')
# plt.scatter(num_steps_list[:limit], e2d2_fidelities[:limit], marker="x")
plt.xlabel("single-qubit depolarising probability")
plt.ylabel("state infidelity")
plt.xscale("log")
plt.yscale("log")
plt.grid(visible=True,which="minor",axis="both")
plt.ylim(bottom=2.5 * 1e-2,top=4.5 * 1e-1)
plt.xticks([1.0 * 1e-5, 3.0 * 1e-5, 1.0 * 1e-4, 3.0 * 1e-4,],
           [r"$10^{-5}$",r"$3\times10^{-5}$",r"$10^{-4}$",r"$3\times10^{-4}$",],
           minor=True)
plt.yticks([3.0 * 1e-2, 1e-1, 3.0*1e-1,],
           [r"$3\times10^{-2}$",r"$10^{-1}$",r"$3\times10^{-1}$"],
           minor=True)
plt.legend([p1, p2], ["Trotter", "proposed"]) # , "shallow encoding and shallow decoding"))
plt.show()
# plt.title("The effect of encoding and decoding strategy")
```
