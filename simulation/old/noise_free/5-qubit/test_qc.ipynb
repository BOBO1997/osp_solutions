```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from pprint import pprint
import pickle
import time
import datetime
```

```python
from qiskit.circuit import QuantumCircuit, Parameter
import qiskit.quantum_info as qi
```

```python
from osp_solutions.circuits_path import gate_path_conventional, gate_path_proposed
from osp_solutions.simulator_dm import compute_distance_trace_unitary_mod_phase
from osp_solutions.hamiltonian import make_H_Heisenberg_path
```

```python
num_qubits = 5
num_steps = 2
time_evolution = np.pi

qc = QuantumCircuit(num_qubits)
qc.compose(gate_path_conventional(num_qubits=num_qubits,
                                  num_steps=num_steps,
                                  dt=Parameter(r"$\Delta t$"), # time_evolution / num_steps,
                                  type_block="3cnot",
                                  to_instruction=False,
                                  add_barrier=True,
                                 ),
           qubits=list(range(num_qubits)),
           inplace=True,
          )
qc.draw("mpl")
```
